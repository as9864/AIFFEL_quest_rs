# 2) 데이터로더 구성
실습 때 다루었던 IMDB 데이터셋은 텍스트를 가공하여 imdb.data_loader() 메서드를 호출하면 숫자 인덱스로 변환된 텍스트와 word_to_index 딕셔너리까지 친절하게 제공합니다. 그러나 이번에 다루게 될 nsmc 데이터셋은 전혀 가공되지 않은 텍스트 파일로 이루어져 있습니다. 이것을 읽어서 imdb.data_loader()와 동일하게 동작하는 자신만의 data_loader를 만들어 보는 것으로 시작합니다. data_loader 안에서는 다음을 수행해야 합니다.

데이터의 중복 제거
NaN 결측치 제거
한국어 토크나이저로 토큰화
불용어(Stopwords) 제거
사전word_to_index 구성
텍스트 스트링을 사전 인덱스 스트링으로 변환
X_train, y_train, X_test, y_test, word_to_index 리턴

```python
from konlpy.tag import Mecab
import numpy as np
from collections import Counter

# 형태소 분석기(Mecab) 초기화
tokenizer = Mecab()

# 분석에서 제외할 불용어(Stopwords) 리스트 정의
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']

def load_data(train_data, test_data, num_words=10000):
    # ----------------------------------------------------
    # 1. 데이터 중복 및 결측치(NaN) 제거
    # ----------------------------------------------------
    # Train 데이터의 리뷰 내용('document' 컬럼) 중복 제거
    train_data.drop_duplicates(subset=['document'], inplace=True)
    # Train 데이터 중 빈 값이 있는 행(Row) 제거
    train_data = train_data.dropna(how = 'any')
    
    # Test 데이터의 리뷰 내용('document' 컬럼) 중복 제거
    test_data.drop_duplicates(subset=['document'], inplace=True)
    # Test 데이터 중 빈 값이 있는 행(Row) 제거
    test_data = test_data.dropna(how = 'any')

    # ----------------------------------------------------
    # 2. Train 데이터 토큰화 및 불용어 제거
    # ----------------------------------------------------
    X_train = []
    for sentence in train_data['document']:
        temp_X = tokenizer.morphs(sentence) # 문장을 형태소 단위로 토큰화
        temp_X = [word for word in temp_X if not word in stopwords] # 불용어에 포함되지 않은 단어만 남김
        X_train.append(temp_X)

    # ----------------------------------------------------
    # 3. Test 데이터 토큰화 및 불용어 제거
    # ----------------------------------------------------
    X_test = []
    for sentence in test_data['document']:
        temp_X = tokenizer.morphs(sentence) # 문장을 형태소 단위로 토큰화
        temp_X = [word for word in temp_X if not word in stopwords] # 불용어에 포함되지 않은 단어만 남김
        X_test.append(temp_X)

    # ----------------------------------------------------
    # 4. 단어 사전(Vocabulary) 구축
    # ----------------------------------------------------
    # 모든 Train 토큰들을 하나의 리스트로 통합
    words = np.concatenate(X_train).tolist()
    # 단어별 등장 빈도수 계산
    counter = Counter(words)
    # 가장 빈도수가 높은 상위 (num_words - 4)개의 단어만 추출
    counter = counter.most_common(num_words - 4)
    
    # 특수 토큰(<PAD>, <BOS>, <UNK>, <UNUSED>)이 들어갈 빈자리 4개를 앞에 확보하고 단어 리스트 생성
    vocab = ['', '', '', ''] + [key for key, _ in counter]
    # 단어를 key로, 인덱스 번호를 value로 갖는 딕셔너리 생성
    word_to_index = {word:index for index, word in enumerate(vocab)}

    # ----------------------------------------------------
    # 5. 텍스트 데이터를 정수 인덱스 벡터로 변환
    # ----------------------------------------------------
    # 중첩 함수: 텍스트 리스트를 변환 규칙에 따라 인덱스 리스트로 바꾸는 함수
    def wordlist_to_indexlist(wordlist):
        # 사전에 있는 단어는 해당 인덱스로, 없는 단어(OOV)는 공백('') 인덱스로 매핑
        # (주의: 보통 공백 자리에 '<UNK>' 같은 특수 토큰 키를 명시적으로 넣는 것이 좋습니다)
        return [word_to_index[word] if word in word_to_index else word_to_index[''] for word in wordlist]

    # Train/Test 전체 데이터에 위 변환 함수를 적용 (Map)
    X_train = list(map(wordlist_to_indexlist, X_train))
    X_test = list(map(wordlist_to_indexlist, X_test))

    # 전처리가 완료된 X(텍스트 인덱스), y(레이블 정답 배열), word_to_index 사전을 반환
    return X_train, np.array(list(train_data['label'])), X_test, np.array(list(test_data['label'])), word_to_index

# 함수를 실행하여 전처리된 데이터셋 할당
X_train, y_train, X_test, y_test, word_to_index = load_data(train_data, test_data)
```

```python
index_to_word = {index:word for word, index in word_to_index.items()}
```

```python
# 문장 1개를 활용할 딕셔너리와 함께 주면, 단어 인덱스 리스트 벡터로 변환해 주는 함수입니다.
# 단, 모든 문장은 <BOS>로 시작하는 것으로 합니다.
def get_encoded_sentence(sentence, word_to_index):
    return [word_to_index['<BOS>']]+[word_to_index[word] if word in word_to_index else word_to_index['<UNK>'] for word in sentence.split()]

# 여러 개의 문장 리스트를 한꺼번에 단어 인덱스 리스트 벡터로 encode해 주는 함수입니다.
def get_encoded_sentences(sentences, word_to_index):
    return [get_encoded_sentence(sentence, word_to_index) for sentence in sentences]

# 숫자 벡터로 encode된 문장을 원래대로 decode하는 함수입니다.
def get_decoded_sentence(encoded_sentence, index_to_word):
    return ' '.join(index_to_word[index] if index in index_to_word else '<UNK>' for index in encoded_sentence[1:])  #[1:]를 통해 <BOS>를 제외

# 여러 개의 숫자 벡터로 encode된 문장을 한꺼번에 원래대로 decode하는 함수입니다.
def get_decoded_sentences(encoded_sentences, index_to_word):
    return [get_decoded_sentence(encoded_sentence, index_to_word) for encoded_sentence in encoded_sentences]
```

# 3) 모델 구성을 위한 데이터 분석 및 가공  
데이터셋 내 문장 길이 분포  
적절한 최대 문장 길이 지정  
torch.nn.utils.rnn.pad_sequence 을 활용한 패딩 추가  


```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 각 문장의 길이 계산 (이미 정수 인코딩된 리스트라고 가정)
num_tokens = [len(tokens) for tokens in X_train]
num_tokens = np.array(num_tokens)

# 2. 통계치 출력
print('문장 길이 최대 : ', np.max(num_tokens))
print('문장 길이 최소 : ', np.min(num_tokens))
print('문장 길이 평균 : ', np.mean(num_tokens))
print('문장 길이 표준편차 : ', np.std(num_tokens))

# 3. 시각화 (히스토그램)
plt.hist(num_tokens, bins=50)
plt.xlabel('Length of samples')
plt.ylabel('Number of samples')
plt.show()
```

```python
# 전체 샘플 중 수용 가능한 비율 설정 (예: 95%)
max_len_ratio = 0.95

# 지정한 비율을 만족하는 임계 길이 계산
max_len = int(np.percentile(num_tokens, max_len_ratio * 100))
print(f'전체 샘플의 {max_len_ratio * 100}%가 포함되는 최대 문장 길이는: {max_len} 입니다.')

# (선택) 실제로 데이터가 얼마나 포함되는지 검증해보기
def below_threshold_len(max_len, nested_list):
    cnt = 0
    for s in nested_list:
        if len(s) <= max_len:
            cnt = cnt + 1
    print(f'전체 샘ple 중 길이가 {max_len} 이하인 샘플의 비율: {(cnt / len(nested_list)) * 100:.2f}%')

below_threshold_len(max_len, X_train)
```

3. torch.nn.utils.rnn.pad_sequence를 활용한 패딩 추가
TensorFlow/Keras의 pad_sequences와 달리, PyTorch의 pad_sequence는 Tensor들의 리스트를 입력으로 받으며, 기본적으로 (가장 긴 문장 길이, 배치 크기) 형태로 패딩을 수행합니다.

우리가 흔히 쓰는 (배치 크기, 문장 길이) 형태로 만들려면 batch_first=True 옵션을 반드시 주어야 합니다. 또한 앞서 구한 max_len으로 길이를 제한하는 슬라이싱 작업도 함께 진행해야 합니다.

```python
import torch
from torch.nn.utils.rnn import pad_sequence

# 1. 계산해 둔 max_len을 기준으로 문장 잘라내기 (Truncation)
# pad_sequence는 길이를 늘려주기만 하므로, max_len보다 긴 문장은 미리 잘라야 합니다.
X_train_trimmed = [torch.tensor(tokens[:max_len]) for tokens in X_train]
X_test_trimmed = [torch.tensor(tokens[:max_len]) for tokens in X_test]

# 2. pad_sequence를 이용해 패딩 진행
# word_to_index에 정의된 '<PAD>' 토큰의 인덱스를 padding_value에 넣어줍니다. (보통 0)
pad_idx = word_to_index['<PAD>'] if '<PAD>' in word_to_index else 0

X_train_padded = pad_sequence(X_train_trimmed, batch_first=True, padding_value=pad_idx)
X_test_padded = pad_sequence(X_test_trimmed, batch_first=True, padding_value=pad_idx)

# 3. 최종 형태(Shape) 확인
print("X_train 변환 후 구조:", X_train_padded.shape)
print("X_test 변환 후 구조:", X_test_padded.shape)
```

# 4) 모델 구성 및 validation set 구성
모델은 3가지 이상 다양하게 구성하여 실험해 보세요.

```python
import torch
from torch.utils.data import TensorDataset, DataLoader

# 1. 정답(Label) 데이터 Tensor 변환 (기존 y_train, y_test 변수 사용)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32)

# 2. Validation Set 분리
num_train = 100000

X_train_split = X_train_padded[:num_train]
y_train_split = y_train_tensor[:num_train]

X_val = X_train_padded[num_train:]
y_val = y_train_tensor[num_train:]

# 3. DataLoader 구축
BATCH_SIZE = 64

train_loader = DataLoader(TensorDataset(X_train_split, y_train_split), batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(TensorDataset(X_val, y_val), batch_size=BATCH_SIZE, shuffle=False)
test_loader = DataLoader(TensorDataset(X_test_padded, y_test_tensor), batch_size=BATCH_SIZE, shuffle=False)

print(f"Train 크기: {len(X_train_split)} | Val 크기: {len(X_val)} | Test 크기: {len(X_test_padded)}")
```

## 2. 하이퍼파라미터 및 3가지 모델 정의 (RNN, CNN, LSTM)
공통적으로 사용할 하이퍼파라미터를 상단에 정의하고, 순서대로 3개 모델 클래스를 선언합니다.

```python
import torch.nn as nn
import torch.nn.functional as F

#vocab_size = len(word_to_index)  # 단어 사전 크기 gpu 에러 발생

# embedding_dim = 100              # 임베딩 벡터 차원
# hidden_dim = 128                 # 은닉층 차원
# 기존: vocab_size = len(word_to_index)
# 변경: 데이터 내 실제 최댓값을 찾아서 안전하게 설정 (오류 방지)
vocab_size = int(torch.max(X_train_padded).item()) + 1

embedding_dim = 100              
hidden_dim = 128                 

print(f"안전하게 재설정된 vocab_size: {vocab_size}")
```

① RNN 모델 (바닐라 RNN)
가장 기본적인 순순형 신경망 구조입니다. 긴 문장에서는 기울기 소실(Vanishing Gradient) 문제가 있을 수 있어 베이스라인으로 적합합니다.

```python
class VanillaRNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim):
        super(VanillaRNN, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=pad_idx)
        self.rnn = nn.RNN(embedding_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, 1)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        embedded = self.embedding(x)
        # rnn_out: (batch, seq_len, hidden_dim)
        rnn_out, hidden = self.rnn(embedded)
        # 마지막 타임스텝의 hidden state 추출
        last_out = rnn_out[:, -1, :] 
        return self.sigmoid(self.fc(last_out)).squeeze()
```

② LSTM 모델 (단방향 LSTM)
RNN의 단점을 보완하여 정보의 장기 의존성을 잘 잡아내는 구조입니다. 텍스트 감성 분류에서 안정적인 성능을 보여줍니다.

```python
class VanillaLSTM(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim):
        super(VanillaLSTM, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=pad_idx)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, 1)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        embedded = self.embedding(x)
        # lstm_out: (batch, seq_len, hidden_dim)
        lstm_out, (hidden, cell) = self.lstm(embedded)
        # 마지막 타임스텝의 hidden state 추출
        last_out = lstm_out[:, -1, :]
        return self.sigmoid(self.fc(last_out)).squeeze()
```

③ CNN 모델 (1D CNN)
문장 내의 인접한 단어들의 특징(N-gram)을 컨볼루션 필터로 포착하는 구조입니다. 병렬 연산이 가능해 RNN 계열보다 학습 속도가 엄청나게 빠릅니다.

```python
class TextCNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim):
        super(TextCNN, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=pad_idx)
        # kernel_size=3 (3개 단어씩 묶어서 특징 추출)
        self.conv = nn.Conv1d(in_channels=embedding_dim, out_channels=hidden_dim, kernel_size=3)
        self.pool = nn.AdaptiveMaxPool1d(1) # 문장의 핵심 특징만 뽑아냄
        self.fc = nn.Linear(hidden_dim, 1)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        embedded = self.embedding(x) # (batch, seq_len, emb_dim)
        # Conv1d는 (batch, channel, seq_len) 형태를 받으므로 차원 축 변경
        embedded = embedded.permute(0, 2, 1) 
        
        conv_out = F.relu(self.conv(embedded))
        pooled = self.pool(conv_out).squeeze(-1) # (batch, hidden_dim)
        return self.sigmoid(self.fc(pooled)).squeeze()
```

3. 학습 및 검증용 공통 루프 함수  
코드가 길어지는 것을 방지하기 위해, 하나의 함수로 모델만 바꿔가며 학습할 수 있도록 루프를 설계했습니다.

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"현재 사용 중인 장치: {device}")

def train_model(model, train_loader, val_loader, epochs=5):
    model = model.to(device)
    criterion = nn.BCELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
    
    for epoch in range(epochs):
        # --- Train ---
        model.train()
        train_loss, train_correct = 0, 0
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item() * inputs.size(0)
            preds = (outputs >= 0.5).float()
            train_correct += (preds == labels).sum().item()
            
        # --- Validation ---
        model.eval()
        val_loss, val_correct = 0, 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                
                val_loss += loss.item() * inputs.size(0)
                preds = (outputs >= 0.5).float()
                val_correct += (preds == labels).sum().item()
                
        print(f"Epoch {epoch+1}/{epochs} | "
              f"Train Loss: {train_loss/len(train_loader.dataset):.4f} Acc: {train_correct/len(train_loader.dataset):.4f} | "
              f"Val Loss: {val_loss/len(val_loader.dataset):.4f} Acc: {val_correct/len(val_loader.dataset):.4f}")
```

# 5) 모델 훈련 개시

1. Vanilla RNN 모델 훈련  
가장 기본적인 RNN 모델을 먼저 생성하고 학습을 시작합니다. 3에포크(Epochs) 정도 가볍게 돌려봅니다.

만약 Train Loss는 계속 떨어지는데 Val Loss가 다시 치솟거나 Val Acc가 정체된다면 과적합이 일어나는 중입니다. (지금은 3에포크만 돌리기 때문에 안전할 것입니다.)

# 6) Loss, Accuracy 그래프 시각화

```python
#Step 1. 기록 기능이 추가된 학습 함수 정의 (새로운 셀)

def train_model_with_history(model, train_loader, val_loader, epochs=3):
    model = model.to(device)
    criterion = nn.BCELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
    
    # 그래프 시각화를 위한 기록용 딕셔너리
    history = {
        'train_loss': [], 'train_acc': [],
        'val_loss': [], 'val_acc': []
    }
    
    for epoch in range(epochs):
        # --- Train ---
        model.train()
        train_loss, train_correct = 0, 0
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item() * inputs.size(0)
            preds = (outputs >= 0.5).float()
            train_correct += (preds == labels).sum().item()
            
        # --- Validation ---
        model.eval()
        val_loss, val_correct = 0, 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                
                val_loss += loss.item() * inputs.size(0)
                preds = (outputs >= 0.5).float()
                val_correct += (preds == labels).sum().item()
        
        # 에포크별 평균값 계산 및 기록
        epoch_train_loss = train_loss / len(train_loader.dataset)
        epoch_train_acc = train_correct / len(train_loader.dataset)
        epoch_val_loss = val_loss / len(val_loader.dataset)
        epoch_val_acc = val_correct / len(val_loader.dataset)
        
        history['train_loss'].append(epoch_train_loss)
        history['train_acc'].append(epoch_train_acc)
        history['val_loss'].append(epoch_val_loss)
        history['val_acc'].append(epoch_val_acc)
                
        print(f"Epoch {epoch+1}/{epochs} | "
              f"Train Loss: {epoch_train_loss:.4f} Acc: {epoch_train_acc:.4f} | "
              f"Val Loss: {epoch_val_loss:.4f} Acc: {epoch_val_acc:.4f}")
              
    return history
```

```python
# Step 2. 모델 재훈련 및 History 받아오기 (새로운 셀)
# 새 모델 인스턴스 생성
cnn_model = VanillaRNN(vocab_size, embedding_dim, hidden_dim)

# 이력을 반환하는 학습 함수로 훈련 진행 (예: 3~5 에포크)
cnn_history = train_model_with_history(cnn_model, train_loader, val_loader, epochs=3)
```

```python
#Step 3. Loss & Accuracy 그래프 시각화 (새로운 셀)
import matplotlib.pyplot as plt

epochs_range = range(1, len(cnn_history['train_loss']) + 1)

plt.figure(figsize=(12, 4))

# 1. Loss 그래프
plt.subplot(1, 2, 1)
plt.plot(epochs_range, cnn_history['train_loss'], 'bo-', label='Training Loss')
plt.plot(epochs_range, cnn_history['val_loss'], 'ro-', label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# 2. Accuracy 그래프
plt.subplot(1, 2, 2)
plt.plot(epochs_range, cnn_history['train_acc'], 'bo-', label='Training Acc')
plt.plot(epochs_range, cnn_history['val_acc'], 'ro-', label='Validation Acc')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Step 2. 모델 재훈련 및 History 받아오기 (새로운 셀)
# 새 모델 인스턴스 생성
cnn_model = VanillaLSTM(vocab_size, embedding_dim, hidden_dim)

# 이력을 반환하는 학습 함수로 훈련 진행 (예: 3~5 에포크)
cnn_history = train_model_with_history(cnn_model, train_loader, val_loader, epochs=3)
```

그래프 해석
정상적인 그래프: 에포크가 진행될수록 파란색(Train)과 빨간색(Val) 선이 함께 아래로(Loss) 또는 위로(Acc) 부드럽게 동행합니다.

과적합(Overfitting) 발생 징후: 파란색 선(Train Loss)은 계속 떨어지는데, 빨간색 선(Val Loss)이 특정 순간부터 다시 꺾여서 위로 올라가기 시작하면 모델이 훈련 데이터에만 너무 과하게 적응했다는 뜻입니다! 다음 단계인 한국어 Word2Vec을 쓰면 이 간격을 좁히는 데 큰 도움이 됩니다.

2. 자체 학습한 임베딩의 단어 유사도 확인하기
Gensim 라이브러리를 사용해 방금 저장한 텍스트 파일을 불러온 뒤, 특정 단어(예: '영화', '최고', '재미' 등)와 가장 유사한 단어들을 찾아봅니다. 3에포크만 학습했기 때문에 생각보다 연관성이 떨어지는 단어가 나올 수 있습니다.

```python
from gensim.models import Word2Vec
import os

# 1. 모델 로드
our_vectors = Word2Vec.load(os.getenv("HOME") + '/data/word2vec_ko.model')

# 테스트해볼 단어 선택
test_word = '영화'

# 2. 단어 검색 및 유사도 추출 시 반드시 .wv 를 거쳐야 합니다!
if test_word in our_vectors.wv:
    print(f" '[{test_word}]'와 유사한 단어 top 5:")
    for word, similarity in our_vectors.wv.most_similar(test_word, topn=5):
        print(f"{word}: {similarity:.4f}")
else:
    print(f"'{test_word}' 단어가 사전에 없습니다.")
```
