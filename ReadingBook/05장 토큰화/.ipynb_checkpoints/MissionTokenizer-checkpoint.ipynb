```python
import re
from collections import Counter

# ==========================================
# 1. 파일 불러오기 및 전처리 (Preprocessing)
# ==========================================
file_path = './datasets/06TheAvengers.txt'  # 코랩 경로에 업로드된 파일 이름

try:
    with open(file_path, 'r', encoding='utf-8') as f:
        raw_text = f.read()
except FileNotFoundError:
    print(f"[오류] {file_path} 파일을 찾을 수 없습니다. 코랩 좌측 파일 탭에 업로드했는지 확인해주세요.")
    # 테스트용 대포 텍스트 (파일이 없을 경우 대비)
    raw_text = "The Tesseract has awakened. It is on a neutral world, a human world."

```

```python
# 1) 모든 문자를 소문자로 변환
lowercase_text = raw_text.lower()
print(lowercase_text)
```

```python

# 2) 모든 특수기호 제거 (알파벳 소문자와 공백만 남김)
# re.sub를 사용하여 알파벳과 공백이 아닌 문자를 제거합니다.
clean_text = re.sub(r'[^a-z\s]', '', lowercase_text)
print(clean_text)
```

```python

# 3) 단어 단위 분리 (띄어쓰기 기준 토큰화)
# split()은 연속된 공백이나 줄바꿈을 자동으로 하나로 처리해 줍니다.
tokens = clean_text.split()

print("--- 전처리 및 토큰화 완료 예시 ---")
print(f"전체 토큰 수: {len(tokens)}")
print(f"처음 10개 토큰: {tokens[:10]}\n")
```

```python


# ==========================================
# 2. 빈도수 딕셔너리(dic) 및 단어 사전(vocab) 구축
# ==========================================
# 1) 각 단어별 빈도수를 딕셔너리 형태로 저장
dic = dict(Counter(tokens))

dic
```

```python

# 2) 빈도수 순으로 내림차순 정렬
# Counter.most_common()을 쓰거나 sorted를 활용합니다.
sorted_words = sorted(dic.items(), key=lambda x: x[1], reverse=True)

sorted_words
```

```python

# 3) 정렬 순서대로 정수 인덱스 부여하여 vocab 생성
vocab = {}
for index, (word, frequency) in enumerate(sorted_words):
    vocab[word] = index
vocab
```

```python

# 4) 가장 뒷부분에 UNK 토큰 추가
# 현재 vocab의 길이 자체가 다음 인덱스 번호가 됩니다.
vocab['<UNK>'] = len(vocab)

vocab['<UNK>']
```

```python

print("--- 빈도수 상위 5개 단어와 인덱스 ---")
for word in list(vocab.keys())[:5]:
    print(f"단어: {word:10} | 인덱스: {vocab[word]}")
print(f"특수 토큰: <UNK>   | 인덱스: {vocab['<UNK>']}\n")


# ==========================================
# 3. 정수를 리턴하는 encoder 함수 구현
# ==========================================
def encoder(input_text):
    """
    입력받은 텍스트를 소문자화, 특수문자 제거, 띄어쓰기 토큰화를 거쳐
    vocab의 정수 인덱스 시퀀스(리스트)로 변환합니다.
    사전에 없는 단어(OOV)는 <UNK> 토큰의 인덱스로 변환합니다.
    """
    # 1) 입력 텍스트 전처리 및 토큰화
    text_lower = input_text.lower()
    text_clean = re.sub(r'[^a-z\s]', '', text_lower)
    input_tokens = text_clean.split()
    
    # 2) Vocab을 참조하여 정수 인덱스로 인코딩 (OOV 대응)
    encoded_sequence = []
    unk_index = vocab['<UNK>']  # UNK 토큰의 정수 인덱스
    
    for token in input_tokens:
        # 단어가 사전에 있으면 해당 인덱스를, 없으면 unk_index를 반환
        encoded_sequence.append(vocab.get(token, unk_index))
        
    return encoded_sequence

# ==========================================
# 4. 인코더 테스트 및 사용 예시
# ==========================================
print("--- 인코더 테스트 실행 ---")
# 예시 1: 사전에 있을 법한 단어들로 구성된 문장
sample_input_1 = "The Avengers saved my life."
result_1 = encoder(sample_input_1)
print(f"입력: '{sample_input_1}'")
print(f"출력: {result_1}\n")

# 예시 2: 사전에 절대 없을 신조어나 외계어(OOV)가 포함된 문장
# 'generative'와 'ops'는 영화 대본에 없을 확률이 높으므로 <UNK> 인덱스로 변환됩니다.
sample_input_2 = "Generative AI and MLOps are awesome!"
result_2 = encoder(sample_input_2)
print(f"입력: '{sample_input_2}'")
print(f"출력: {result_2}  <- 사전에 없는 단어는 맨 마지막 인덱스({vocab['<UNK>']})로 매핑됨")
```
